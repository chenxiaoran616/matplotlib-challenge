```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

mouse_data_file="data/mouse_drug_data.csv"
clinical_data_file="data/clinicaltrial_data.csv"

mouse_data_df=pd.read_csv(mouse_data_file)
clinical_data_df=pd.read_csv(clinical_data_file)

# combine two files into one
test_data_df=pd.merge(clinical_data_df,mouse_data_df,on="Mouse ID",how="outer")
test_data_df.head()
```

# Tumor Response to Treatment

```python
#group the data by drug and timepoint and show mean
drug_effect_mean0 = test_data_df.groupby(["Drug","Timepoint"]).mean()["Tumor Volume (mm3)"]
drug_effect_mean=pd.DataFrame(drug_effect_mean0)
drug_effect_mean=drug_effect_mean.reset_index()
drug_effect_mean.head()
```

```python
#group the data by drug and timepoint and store the standard error
drug_effect_sem0 = test_data_df.groupby(["Drug","Timepoint"]).sem()["Tumor Volume (mm3)"]
drug_effect_sem=pd.DataFrame(drug_effect_sem0)
drug_effect_sem=drug_effect_sem.reset_index()
drug_effect_sem.head()
```

```python
mean_pivot=drug_effect_mean.pivot(index="Timepoint", columns="Drug")["Tumor Volume (mm3)"]
mean_pivot_df=pd.DataFrame(mean_pivot)
mean_pivot_df
```

```python
sem_pivot=drug_effect_sem.pivot(index="Timepoint",columns="Drug")["Tumor Volume (mm3)"]
sem_pivot_df=pd.DataFrame(sem_pivot)
sem_pivot_df
```

```python
plt.errorbar(mean_pivot_df.index, mean_pivot_df["Capomulin"], yerr=sem_pivot_df["Capomulin"], color="red", marker="o", markersize=5, linestyle="dashed", linewidth=0.50,label="Capomulin")
plt.errorbar(mean_pivot_df.index, mean_pivot_df["Infubinol"], yerr=sem_pivot_df["Infubinol"], color="green", marker="o", markersize=5, linestyle="dashed", linewidth=0.50,label="Infubinol")
plt.errorbar(mean_pivot_df.index, mean_pivot_df["Ketapril"], yerr=sem_pivot_df["Ketapril"], color="blue", marker="o", markersize=5, linestyle="dashed", linewidth=0.50,label="Ketapril")
plt.errorbar(mean_pivot_df.index, mean_pivot_df["Placebo"], yerr=sem_pivot_df["Placebo"], color="black", marker="o", markersize=5, linestyle="dashed", linewidth=0.50, label="Placebo")
plt.ylabel("Tumor Volume(mm3)")
plt.xlabel("Timepoint")
plt.title("Tumor Response to Treatment")
plt.legend()
plt.grid()
plt.savefig("Fig1.png")
plt.show
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 

drug_effect_mean0 = test_data_df.groupby(["Drug","Timepoint"]).mean()["Metastatic Sites"]
# Convert to DataFrame
metastatic_mean_df=pd.DataFrame(drug_effect_mean0)
# Preview DataFrame
metastatic_mean_df=metastatic_mean_df.reset_index()
metastatic_mean_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
metastatic_sem0 = test_data_df.groupby(["Drug","Timepoint"]).sem()["Metastatic Sites"]
metastatic_sem_df=pd.DataFrame(metastatic_sem0)
metastatic_sem_df=metastatic_sem_df.reset_index()
metastatic_sem_df.head()
# Convert to DataFrame
```

```python
# Minor Data Munging to Re-Format the Data Frames
metastatic_mean_pivot=metastatic_mean_df.pivot(index="Timepoint", columns="Drug")["Metastatic Sites"]
metastatic_mean_pivot_df=pd.DataFrame(metastatic_mean_pivot)
# Preview that Reformatting worked
metastatic_mean_pivot_df.head()
```

```python
metastatic_sem_pivot=metastatic_sem_df.pivot(index="Timepoint",columns="Drug")["Metastatic Sites"]
metastatic_sem_pivot_df=pd.DataFrame(metastatic_sem_pivot)
metastatic_sem_pivot_df.head()
```

```python
plt.errorbar(metastatic_mean_pivot_df.index, metastatic_mean_pivot_df["Capomulin"], yerr=sem_pivot_df["Capomulin"], color="red", marker="o", markersize=5, linestyle="dashed", linewidth=0.50,label="Capomulin")
plt.errorbar(metastatic_mean_pivot_df.index, metastatic_mean_pivot_df["Infubinol"], yerr=sem_pivot_df["Infubinol"], color="green", marker="o", markersize=5, linestyle="dashed", linewidth=0.50,label="Infubinol")
plt.errorbar(metastatic_mean_pivot_df.index, metastatic_mean_pivot_df["Ketapril"], yerr=sem_pivot_df["Ketapril"], color="blue", marker="o", markersize=5, linestyle="dashed", linewidth=0.50,label="Ketapril")
plt.errorbar(metastatic_mean_pivot_df.index, metastatic_mean_pivot_df["Placebo"], yerr=sem_pivot_df["Placebo"], color="black", marker="o", markersize=5, linestyle="dashed", linewidth=0.50, label="Placebo")
plt.ylabel("Met. Sites")
plt.xlabel("Treatment Duration(Days)")
plt.title("Metastatic Spread During Treatment")
plt.grid()
plt.legend()
plt.savefig("fig2")
plt.show
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
count_of_mice=test_data_df.groupby(["Drug","Timepoint"]).count()["Tumor Volume (mm3)"]

# Convert to DataFrame
count_of_mice_df=pd.DataFrame(count_of_mice)
count_of_mice_df=count_of_mice_df.reset_index()
#count_of_mice_df=count_of_mice_df.rename(columns={"Tumor Volume(mm3)":"Mouse Count"})
count_of_mice_df.columns=["Drug","Timepoint","Mouse Count"]
count_of_mice_df.head()
# Preview DataFrame
```

```python
count_mice_pivot=count_of_mice_df.pivot(index="Timepoint",columns="Drug")["Mouse Count"]
count_mice_pivot
```

```python
# Minor Data Munging to Re-Format the Data Frames
plt.errorbar(count_mice_pivot.index, count_mice_pivot["Capomulin"], yerr=sem_pivot_df["Capomulin"], color="red", marker="o", markersize=5, linestyle="dashed", linewidth=0.50,label="Capomulin")
plt.errorbar(count_mice_pivot.index, count_mice_pivot["Infubinol"], yerr=sem_pivot_df["Infubinol"], color="green", marker="o", markersize=5, linestyle="dashed", linewidth=0.50,label="Infubinol")
plt.errorbar(count_mice_pivot.index, count_mice_pivot["Ketapril"], yerr=sem_pivot_df["Ketapril"], color="blue", marker="o", markersize=5, linestyle="dashed", linewidth=0.50,label="Ketapril")
plt.errorbar(count_mice_pivot.index, count_mice_pivot["Placebo"], yerr=sem_pivot_df["Placebo"], color="black", marker="o", markersize=5, linestyle="dashed", linewidth=0.50, label="Placebo")
plt.ylabel("Suvival Rate")
plt.xlabel("Time(Days)")
plt.title("Survival Rate During Treatment")
plt.grid()
plt.legend()
plt.savefig("Fig3")
plt.show
# Preview the Data Frame
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
tumor_percentage_change=100*(mean_pivot_df.iloc[-1]-mean_pivot_df.iloc[0])/mean_pivot_df.iloc[0]
sem_percentage_change=100*(sem_pivot_df.iloc[-1]-sem_pivot_df.iloc[0])/sem_pivot_df.iloc[0]
tumor_percentage_change
# Display the data to confirm
```

```python
percent_changes=(tumor_percentage_change["Capomulin"],
                tumor_percentage_change["Infubinol"],
                tumor_percentage_change["Ketapril"],
                tumor_percentage_change["Placebo"])

fig, ax = plt.subplots()
ind = np.arange(len(percent_changes))  
width = 1
rectsPass = ax.bar(ind[0], percent_changes[0], width, color='green')
rectsFail = ax.bar(ind[1:], percent_changes[1:], width, color='red')

ax.set_ylabel('% Tumor Volume Change')
ax.set_title('Tumor Change Over 45 Day Treatment')
ax.set_xticks(ind + 0.5)
ax.set_xticklabels(('Capomulin', 'Infubinol', 'Ketapril', 'Placebo'))
ax.set_autoscaley_on(False)
ax.set_ylim([-30,70])
ax.grid(True)


def autolabelFail(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 3,
                '%d%%' % int(height),
                ha='center', va='bottom', color="white")
        
def autolabelPass(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., -8,
                '-%d%% ' % int(height),
                ha='center', va='bottom', color="white")

autolabelPass(rectsPass)
autolabelFail(rectsFail)

fig.savefig("Fig4.png")

fig.show()
```

## Observations

1. Capomulin is the only drug that's effective on controlling tumor growth
2. With usage of each of the four drugs, metastatic sites grow over the using period of the drug. It grows the fastest under Placebo, and slowest under Capomulin
3. survival rate decrease overtime for all four drugs. But Capomulin's survival rate decreases the slowest, it tells us this drug is more effective than other three
